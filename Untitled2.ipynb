```python
cohen_csv = pd.read_csv(cohen_csvpath, nrows=None)
#idx_pa = csv["view"] == "PA"  # Keep only the PA view
views = ["PA", "AP", "AP Supine", "AP semi erect", "AP erect"]
cohen_idx_keep = cohen_csv.view.isin(views)
cohen_csv = cohen_csv[cohen_idx_keep]
f1_csv = pd.read_csv(f1_csvpath, encoding='ISO-8859-1', nrows=None)
```

```python
f1_df = pd.DataFrame(columns = ["old_path", 'image_path','class'])

for index in f1_csv.index:
    disease = f1_csv['finding'][index]

    name = f1_csv["patientid"][index]
    if disease == 'COVID-19':
      
        if os.path.exists(os.path.join(f1_imgpath, name + '.jpg')):
                sufix = ".jpg"
        elif os.path.exists(os.path.join(f1_imgpath, name + '.png')):
                sufix = ".png"
                
        f1_df = f1_df.append({"old_path":name + sufix, 'image_path': name,'class': disease}
                         ,ignore_index=True)
        
    elif disease == 'Pneumonia':
        if os.path.exists(os.path.join(f1_imgpath, name + '.jpg')):
                sufix = ".jpg"
        elif os.path.exists(os.path.join(f1_imgpath, name + '.png')):
                sufix = ".png"
                
        f1_df = f1_df.append({"old_path":name + sufix, 'image_path': name,'class': disease}
                         ,ignore_index=True)
```

```python
cohen_df = pd.DataFrame(columns =['patientID', 'image_path', 'class'])

for index in cohen_csv.index:
    f = cohen_csv['finding'][index].split(',')[0] # take the first finding, for the case of COVID-19, ARDS
    if f in mapping: 
        entry = {"patientID":cohen_csv['patientid'][index], "image_path": cohen_csv['filename'][index],
                'class': f}
        cohen_df = cohen_df.append(entry,ignore_index = True)
cohen_df
```
